# Agrupamiento y Sorteo

En este JNB se revisará el tema de Agrupamiento y Sorteo en Pandas.

```python
import pandas as pd
reviews = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-GA-AD22/pandas-ricardo-qm/main/data/winemag-data-130k-v2.csv", index_col=0)
pd.set_option("display.max_rows", 5)
```

```python
reviews
```

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

# Multiíndices

Cuando cada renglón se accede por 2 o más valores de campos, se tiene un multindice.

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
countries_reviewed.reset_index()
```

# Sorteo

Se pueden presentar los datos de un Dataframe de manera ordenada por un cierto criterio

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_index()
```

```python
countries_reviewed.sort_values(by=['country', 'len'])
```
